```python
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt

# load csv files into dataframes
jan = pd.read_csv('../data/USS daily waiting time/download_universal-studios-singapore_jan.csv')
feb = pd.read_csv('../data/USS daily waiting time/download_universal-studios-singapore_feb.csv')
mar = pd.read_csv('../data/USS daily waiting time/download_universal-studios-singapore_mar.csv')
apr = pd.read_csv('../data/USS daily waiting time/download_universal-studios-singapore_apr.csv')
may = pd.read_csv('../data/USS daily waiting time/download_universal-studios-singapore_may.csv')
jun = pd.read_csv('../data/USS daily waiting time/download_universal-studios-singapore_jun.csv')
jul = pd.read_csv('../data/USS daily waiting time/download_universal-studios-singapore_jul.csv')
aug = pd.read_csv('../data/USS daily waiting time/download_universal-studios-singapore_aug.csv')
sep = pd.read_csv('../data/USS daily waiting time/download_universal-studios-singapore_sep.csv')
oct = pd.read_csv('../data/USS daily waiting time/download_universal-studios-singapore_oct.csv')
nov = pd.read_csv('../data/USS daily waiting time/download_universal-studios-singapore_nov.csv')
dec = pd.read_csv('../data/USS daily waiting time/download_universal-studios-singapore_dec.csv')

list_of_df = [jan, feb, mar, apr, may, jun, jul, aug, sep, oct, nov, dec]
print(jan)
```

```python
# Define the start and end dates for each month in 2024
start_dates = [
    "2024-01-01 00:00:00", "2024-02-01 00:00:00", "2024-03-01 00:00:00",
    "2024-04-01 00:00:00", "2024-05-01 00:00:00", "2024-06-01 00:00:00",
    "2024-07-01 00:00:00", "2024-08-01 00:00:00", "2024-09-01 00:00:00",
    "2023-10-01 00:00:00", "2023-11-01 00:00:00", "2023-12-01 00:00:00"
]

end_dates = [
    "2024-01-31 23:59:59", "2024-02-29 23:59:59", "2024-03-31 23:59:59",
    "2024-04-30 23:59:59", "2024-05-31 23:59:59", "2024-06-30 23:59:59",
    "2024-07-31 23:59:59", "2024-08-31 23:59:59", "2024-09-30 23:59:59",
    "2023-10-31 23:59:59", "2023-11-30 23:59:59", "2023-12-31 23:59:59"
]

month_names = ["jan", "feb", "mar", "apr", "may", "jun", "jul", "aug", "sep", "oct", "nov", "dec"]

# Here we will tidying up the df by trimming the days that are not within the time frame intended (1 Nov 2023 - 31 Oct 2024)

filtered_dfs = {}

for df, start_date, end_date, month in zip(list_of_df, start_dates, end_dates, month_names):
    df['Date/Time'] = pd.to_datetime(df['Date/Time']) # Ensure 'Date/Time' is in datetime format
    
    # tidying up the df by trimming the days that are not within the time frame intended
    filtered_df = df[(df['Date/Time'] >= start_date) & (df['Date/Time'] <= end_date)]
    
    filtered_dfs[month] = filtered_df
```

```python
# OPTIONAL CELL (TO CHECK OUTPUT)
for month, month_df in filtered_dfs.items():
    print(f"{month}:\n", month_df)
```

```python
# Reformatting the datas
for month in filtered_dfs.keys():
    filtered_dfs[month]['Month'] = filtered_dfs[month]['Date/Time'].dt.month
    filtered_dfs[month]['Day'] = filtered_dfs[month]['Date/Time'].dt.day
    filtered_dfs[month]['Time'] = filtered_dfs[month]['Date/Time'].dt.strftime('%H:%M')
    filtered_dfs[month]['Day of Week'] = filtered_dfs[month]['Date/Time'].dt.day_name()
    filtered_dfs[month] = filtered_dfs[month].drop(columns=['Park'], errors='ignore') 
```

```python
# OPTIONAL CELL (TO CHECK OUTPUT)

for month, month_df in filtered_dfs.items():
    print(f"{month}:\n", month_df)
```

```python
# Importing CSV file on holiday information within the time frame intended (1 Nov 2023 - 31 Oct 2024)
# CSV file generated by ChatGPT
holidays_df = pd.read_csv('../data/singapore_holidays_oct2023_sep2024.csv')
```

```python
# Setting up filters
school_holidays_mask_1 = holidays_df[(holidays_df['School Holiday'] == 1) & (holidays_df['Public Holiday'] == 0)]
school_holidays_mask_2 = holidays_df[(holidays_df['Public Holiday'] == 1) & (holidays_df['School Holiday'] == 0)]
school_holidays_mask_3 = holidays_df[
    (holidays_df['School Holiday'] == 0) & 
    (holidays_df['Public Holiday'] == 0) & 
    (~holidays_df['Day of the Week'].isin(['Saturday', 'Sunday']))]

school_holidays_mask_4 = holidays_df[
    (holidays_df['School Holiday'] == 0) & 
    (holidays_df['Public Holiday'] == 0) & 
    (holidays_df['Day of the Week'].isin(['Saturday', 'Sunday']))]

school_holidays_mask_5 = holidays_df[(holidays_df['Public Holiday'] == 1) & (holidays_df['School Holiday'] == 1)]
```

```python
# Now we create dictionaries with filtered DataFrames for each type of holiday per month

# Initialize dictionaries to store the results
school_holiday_dfs = {}
public_holiday_dfs = {}
weekday_non_holiday_dfs = {}
weekend_non_holiday_dfs = {}
both_holiday_dfs = {}

# Loop through each month in list_of_months
for month in filtered_dfs.keys():
    school_holiday_dfs[month] = filtered_dfs[month][
        filtered_dfs[month][['Month', 'Day']].apply(tuple, axis=1).isin(
            school_holidays_mask_1[['Month', 'Day']].apply(tuple, axis=1)
        )
    ]

    public_holiday_dfs[month] = filtered_dfs[month][
        filtered_dfs[month][['Month', 'Day']].apply(tuple, axis=1).isin(
            school_holidays_mask_2[['Month', 'Day']].apply(tuple, axis=1)
        )
    ]

    weekday_non_holiday_dfs[month] = filtered_dfs[month][
        filtered_dfs[month][['Month', 'Day']].apply(tuple, axis=1).isin(
            school_holidays_mask_3[['Month', 'Day']].apply(tuple, axis=1)
        )
    ]

    weekend_non_holiday_dfs[month] = filtered_dfs[month][
        filtered_dfs[month][['Month', 'Day']].apply(tuple, axis=1).isin(
            school_holidays_mask_4[['Month', 'Day']].apply(tuple, axis=1)
        )
    ]

    both_holiday_dfs[month] = filtered_dfs[month][
        filtered_dfs[month][['Month', 'Day']].apply(tuple, axis=1).isin(
            school_holidays_mask_5[['Month', 'Day']].apply(tuple, axis=1)
        )
    ]
```

```python
# OPTIONAL CELL (TO CHECK OUTPUT)
print(both_holiday_dfs.keys())
print(both_holiday_dfs['jun'])
```

```python
# Now we use groupby Time to calculate the average wait times in 5 minute intervals

# Initialize dictionaries to store the average wait times
average_wait_times = {
    'school_holiday': {},
    'public_holiday': {},
    'weekday_non_holiday': {},
    'weekend_non_holiday': {},
    'both_holiday': {}
}

# Loop through each dictionary
for holiday_type, dfs in {
    'school_holiday': school_holiday_dfs,
    'public_holiday': public_holiday_dfs,
    'weekday_non_holiday': weekday_non_holiday_dfs,
    'weekend_non_holiday': weekend_non_holiday_dfs,
    'both_holiday': both_holiday_dfs
}.items():
    
    for month, df in dfs.items():
        # Check if the DataFrame is not empty
        if not df.empty:
            # Calculate the average wait time for each time
            avg_wait_time = df.groupby('Time')['Wait Time'].mean().reset_index()
            avg_wait_time.columns = ['Time', 'Average Wait Time']
            
            # Calculate the number of visitors (assuming 5 visitors per minute of wait time)
            avg_wait_time['Number of Visitors'] = (
                avg_wait_time['Average Wait Time'].round() * 5  # Round the wait time and multiply by 5
            )
            # Calculate number of new visitors coming in every 5 minutes
            avg_wait_time['Number of New Visitors Arriving'] = (
                avg_wait_time['Number of Visitors']
                - avg_wait_time['Number of Visitors'].shift(1)
            ) * 40
            # To fill the NaN value in the first row with the initial value
            avg_wait_time['Number of New Visitors Arriving'].iloc[0] = (
                avg_wait_time['Number of Visitors'].iloc[0]
            )

            # Store the result in the dictionary
            average_wait_times[holiday_type][month] = avg_wait_time
```

```python
# OPTIONAL CELL (TO CHECK OUTPUT)
print(average_wait_times['school_holiday']['jun'])
print(average_wait_times['weekend_non_holiday']['apr'])
```

We assume that there is a relation between waiting time and number of visitors coming into the park
We estimate 1 minute of waiting time equals to 5 visitors.
Another assumption is 20% of visitors purchase fast-pass tickets.

```python

# Assuming `average_wait_times['weekend_non_holiday']` is a dictionary of DataFrames for each month
average_wait_times_by_holidays = {}

for holiday_type in average_wait_times.keys():

    # Step 1: Combine the DataFrames for all presented months (one per key: 'jan', 'feb', ..., 'dec')
    across_the_months = pd.DataFrame()

    for month in average_wait_times[holiday_type].keys():
        # Access the DataFrame for the current month
        month_df = average_wait_times[holiday_type][month]
        
        # Add a column to indicate the month (optional but useful for tracking)
        month_df['Month'] = month
        
        # Concatenate the current month's data to the combined DataFrame
        across_the_months = pd.concat([across_the_months, month_df], ignore_index=True)

    # Step 2: Group by 'Time' and calculate the average for all the columns
    average_across_months = across_the_months.groupby('Time').agg({
        'Average Wait Time': 'mean',
        'Number of Visitors': 'mean',
        'Number of New Visitors Arriving': 'mean'
    }).reset_index()

    # Step 3: Add this df into the main dictionary:
    average_wait_times_by_holidays[holiday_type] = average_across_months

```

```python
# Optional, to check output
for holiday_type in average_wait_times.keys():
    print(f"For the tyoe of holiday of {holiday_type}, the df is:")
    print(average_wait_times_by_holidays[holiday_type])
```

```python
# Export DataFrame to CSV file
import os

# Create output directory 
output_dir = f'../data/spawning csv files/'
os.makedirs(output_dir, exist_ok=True)

# Export each DataFrame to a CSV file
for holiday_type, df in average_wait_times_by_holidays.items():
    csv_file = os.path.join(output_dir, f'{holiday_type}.csv')
    df.to_csv(csv_file, index=False) # index=False to avoid writing row numbers
```
